Notebook d'exploration interactif du pipeline.

```python
import sys
sys.path.insert(0, '../src')

from eloquent.config import load_config
from eloquent.providers import build_provider_from_config
from eloquent.prompting import build_strategy
from eloquent.logger import setup_logging

setup_logging()
cfg = load_config('../configs/baseline_groq.yaml')
provider = build_provider_from_config(cfg)
strategy = build_strategy(cfg.prompting.strategy)
```

```python
# Test manuel sur une question
question = "En France, quelle est la durée légale du travail hebdomadaire ?"
messages = strategy.build_messages(question)
resp = provider.generate(messages, temperature=0.0, max_tokens=150)
print(f"Réponse : {resp.content}")
print(f"Latence : {resp.latency_ms:.0f}ms")
```
